```python
import numpy as np
from PIL import Image as pim
```

```python
image = pim.open('input/bin_image.png')
image
```

```python
image = np.array(image)
image.shape
```

```python
padding = (1, 1)
padded_image = np.pad(image, padding, 'edge')
padded_height, padded_width = padded_image.shape
```

```python
padded_height, padded_width
```

```python
rank = 9

filtered_image = image.copy()

for i in range(1, padded_height - 1):
    for j in range(1, padded_width - 1):
        block = padded_image[i - 1 : i + 2, j - 1 : j + 2]
        white_count = np.count_nonzero(block)

        filtered_image[i - 1][j - 1] = (white_count >= rank) * 255
```

```python
pim.fromarray(filtered_image)
```
